** Findings **  
The downloaded dataframes are all in a similar format, so they can be looped through with the same functions.

Of note in the sample shown: there are many NaNs in the data - representing either that data was not available, or that the country did not exist during that timespan. For this project we will filter out any years that contain NaNs.

We will also inspect the Indicator Name and Indicator Code columns, in case there is an abberant value that we should be aware of.

```python
def init_df(file_name):
    """
    Reads the file into a dataframe, check and drop unneeded columns
    
    Inputs:
    file_name: The name of the file to be read (in the same wd as this notebook)
    """
    
    #Read the exel file
    df=pd.read_excel(file_name, sheet_name='Data', header=3)
    
    #Make sure there's nothing needed in the indicator cols
    for col in ['Indicator Name', 'Indicator Code']:
        if len(df[col].unique()) >1:
            response=input(f'Multiple distinct entries found in {col}. Delete? (y/n)\n' )
            print(df[col].unique())
            if response.lower()=='y':
                del df[col]
        else: 
            del df[col]
     
    print('Import complete' )    
    return df


def filter_for_country(df, countries):
    """
    Filter only for countries of interest and return a dataframe
    
    Inputs:
    df: The dataframe to be filtered
    countries: The list of countries to keep, must be in WorldBank format.
    """
    
    #Find all entries where the names match the list of countries    
    df_interest=df[df['Country Name'].isin(countries)]
    
    #If not all countries found (or more than intended), raise an error to alert the user
    if len(countries)!=df_interest.shape[0]:
        raise ValueError("There is a mismatch between filtered countries and input list.")
    
    #Reset index to match new order
    df_interest.reset_index(drop=True, inplace=True)
    
    return df_interest
```

```python
#Run the functions defined above to load and clean data


#These are the exact names used by worldbank
countries= ['China', 'Russian Federation', 'Germany', 'United Kingdom', 'France', 
            'Italy', 'Iran, Islamic Rep.', 'Saudi Arabia', 'Israel','Korea, Rep.']

#Prepare a storage list for the processed dataframes
data=[]

#Loop through the raw data files
for i in raw_data:
    df= init_df(i)
    df=filter_for_country(df, countries)
    data.append(df)   
```

```python
def df_extract_years(df, start_year, end_year):
    """
    Extract the years while keeping the label columns
    """
    #Check input values in case of user error
    if start_year >= end_year:
        raise ValueError('Please ensure start year is less than end year.')
    
    #Create a list of year columns to keep, as well as the ID columns
    years= [str(i) for i in range(start_year, end_year+1)]
    years=['Country Name', 'Country Code']+years
    
    #Filter the existing dataframe to drop all columns not present in the list
    df=df.filter(years)
    
    return df
```

```python
#Loop through the data to extract only the years 2010 to 2017
#These years have 0 NAs for any dataframe, and are easy to reference
data=[df_extract_years(datum,2010, 2017) for datum in data]
```

```python

def line_plot_overview(df, data_type):
    """
    Generate a basic lineplot to get a broad look at the data for all countries and all years
    
    Inputs:
    df : The dataframe to be plotted, containing GDP, Military or Population data
    data_type: The code indicating what type of data is contained, GDP, MIL or POP
    """
    #Check inputs
    if data_type not in ['GDP', 'POP', 'MIL']:
        raise ValueError("Please set data_type to 'GDP', 'POP' or 'MIL'")
    
    #Create a dictionary of strings for dynamic title and axis labeling
    str_dict={'GDP':['Gross Domestic Product', 'GDP in Billions (USD)'], 
              'POP':['Population', 'Population in Millions (Total Estimated)'], 
              'MIL':['Military Expenditure', 'Expenditure in Billions (USD)']}
    
    #Generate figure    
    plt.figure(figsize=(11,4))
    
    #For each country, plot a different coloured line
    for i in range(10):
        plt.plot(df.columns[2:], df.iloc[i,2:])
    
    #Assign titles and axes using the dictionary defined above
    plt.title(str_dict[data_type][0])
    plt.ylabel(str_dict[data_type][1])
    plt.legend(df.iloc[:,0], bbox_to_anchor=(1.0, .9))
```

```python

def global_line_dollars(df_dollars, df_pop, dollar_type, mode='abs'):
    """
    Plots the GDP or Military Expenditure for all countries, as both absolute values and per capita.
    
    Inputs:
    df_dollars : The dataframe to be plotted, containing GDP or Military data
    df_pop : The dataframe to be plotted, containing Population data
    dollar_type: The code indicating what type of data is contained, GDP or MIL
    mode: Indicates if desired output is per capita ('per') or absolute amount ('abs')
    """
  
    #Check inputs
    if dollar_type not in ['GDP', 'MIL']:
        raise ValueError("Please set data_type to 'GDP' or 'MIL'")
        
    #Select dictionary for title and axes assignment, and transform dataframe according to selection    
    if mode =='abs':
        str_dict={'GDP':['GDP', 'GDP in Billions (USD)'], 
          'MIL':['Military Expenditure', 'Expenditure in Billions (USD)']}
            
        #Select only the values of the dataframe
        df_val=df_dollars.iloc[:,2:]
    elif mode =='per':
        str_dict={'GDP':['GDP per Capita', 'GDP per Capita in Billions (USD)'], 
                  'MIL':['Military Expenditure per Capita', 'Expenditure per Capita in Billions (USD)']}
        
        #Select only the values, and divide by the population for each respective year
        df_val=df_dollars.iloc[:,2:]/df_pop.iloc[:,2:]
    
    else:
        raise ValueError("Please set mode to 'abs' or 'per'.")
    

    
    #Initialize plot
    plt.figure(figsize=(11,4))
    #Plot the transformed dataframe - either just values or per capita values
    for i in range(10):
        plt.plot(df_val.columns[:], df_val.iloc[i,:])
    
    #Create titles using the dictionary above
    plt.title(str_dict[dollar_type][0])
    plt.ylabel(str_dict[dollar_type][1])
    plt.legend(df_dollars.iloc[:,0], bbox_to_anchor=(1.0, .9))
    
```

```python

global_line_dollars(data_gdp, data_pop, 'GDP', 'abs')
global_line_dollars(data_gdp, data_pop, 'GDP', 'per')


global_line_dollars(data_mil, data_pop, 'MIL', 'abs')
global_line_dollars(data_mil, data_pop, 'MIL', 'per')
```

```python

def mil_as_percent_gdp(df_gdp, df_mil):
    """
    Plot the Military expenditure as a percentage of GDP for all countries.
    
    Inputs:
    df_gdp : The dataframe to be plotted, containing GDP data
    df_mil : The dataframe to be plotted, containing Military data
    """
    
    #Calculate the proportion of GDP the military expenses reperesents
    df_percent=df_mil.iloc[:,2:]/df_gdp.iloc[:,2:]*100
    
    
    #Initialize the figure
    plt.figure(figsize=(11,4))
    for i in range(10):
        plt.plot(df_percent.columns[:], df_percent.iloc[i,:])
    
    #Assign titles and axis strings - this plot only performs one function so no need for dynamic strings
    plt.title('Military Expenditure as a Fraction of GDP')
    plt.ylabel('% GDP')
    plt.legend(df_mil.iloc[:,0], bbox_to_anchor=(1.0, .9))
    
    plt.show()
```

```python

def single_country_metrics_plot(df_gdp,df_mil, df_pop, country_str, mode='abs'):
    """
    Make a lineplot with two axes to compare GDP and Mil expenditure for one country at a time.
    
    Inputs:
    df_gdp : The dataframe to be plotted, containing GDP data
    df_mil : The dataframe to be plotted, containing Military data
    df_pop : The dataframe to be plotted, containing Population data
    country_str: The name of the country to be plotted, as a string
    mode: Indicates if desired output is per capita ('per') or absolute amount ('abs')
    """

    #Checks if country_str is a valid selection
    if country_str in df_mil['Country Name'].to_list():
        
        #Filter dataframes for the country indicated
        df_mil=df_mil[df_mil['Country Name']==country_str]
        df_gdp=df_gdp[df_gdp['Country Name']==country_str]
        df_pop=df_pop[df_pop['Country Name']==country_str]
        
    else:
        raise ValueError('Country not found, please check spelling.')
    
    #Dynamically select title/axes strings according to mode selection
    if mode=='abs':
        str_add='' #blank string

    elif mode=='per':
        str_add=' per Capita'
        #Transform the dataframes to reflect per capita values
        df_gdp.iloc[:,2:]=df_gdp.iloc[:,2:]/df_pop.iloc[:,2:]    
        df_mil.iloc[:,2:]=df_mil.iloc[:,2:]/df_pop.iloc[:,2:]   
        
    else:
        raise ValueError("Please set mode to 'abs' or 'per'.")        
    
    
    
    #Select colours for the lineplots and axes labels
    colours=('Gray', '#0165fc' )
    
    #Initialize the plot
    fig, ax1 = plt.subplots(figsize=(11,4))
    
    #Ax1 is the GDP plot, with its scale on the left
    #Set the y-labels and ticks to the color of the line to match
    ax1.plot(df_gdp.columns[2:], df_gdp.iloc[0,2:], color=colours[0])
    ax1.set_ylabel(f'GDP{str_add} in Billions (USD)', color=colours[0])
    ax1.tick_params(axis='y', labelcolor=colours[0])
    
    #Ax2 is the MIL plot, scale to the right
    #twinx sets the x-axis to the same as ax1
    ax2 = ax1.twinx()
    #Set the y-labels and ticks to the color of the line to match
    ax2.set_ylabel(f'Military Expenditure{str_add} in Billions (USD)', color=colours[1]) 
    ax2.plot(df_mil.columns[2:], df_mil.iloc[0,2:], color=colours[1])
    ax2.tick_params(axis='y', labelcolor=colours[1])
    
    #Set the title
    plt.title(f'Comparison of GDP{str_add} and Military Expenditure{str_add}: {country_str}', fontsize=16)

    plt.show()


```

```python
single_country_metrics_plot(data_gdp, data_mil,  data_pop, 'France')
single_country_metrics_plot(data_gdp, data_mil,  data_pop, 'France', 'per')
```

Next we'll examine the global total of GDP and Military spending, and what proportion each country is responsible for.

```python

def global_stack_dollars(df_dollars, df_pop, dollar_type, mode='abs'):
    """
    Plot the Military expenditure as a stacked plot, to get a feel of proportional spending.
    
    Inputs:
    df_dollars : The dataframe to be plotted, containing GDP or Military data
    df_pop : The dataframe to be plotted, containing Population data
    dollar_type: The code indicating what type of data is contained, GDP or MIL
    mode: Indicates if desired output is per capita ('per') or absolute amount ('abs')
    """
    
    #Check mode and select string dictionary, transform dataframes
    if mode == 'abs':
        str_dict={'GDP':['GDP', 'GDP in Billions (USD)'], 
          'MIL':['Military Expenditure', 'Expenditure in Billions (USD)']}
            
        #Select only values
        df_val=df_dollars.iloc[:,2:]
            
    elif mode == 'per':
        str_dict={'GDP':['GDP per Capita', 'GDP per Capita in Billions (USD)'], 
                  'MIL':['Military Expenditure per Capita', 'Expenditure per Capita in Billions (USD)']}
        
        #Select values and divide by population in those years
        df_val=df_dollars.iloc[:,2:]/df_pop.iloc[:,2:]
        
    else:
        raise ValueError("Please set mode to 'abs' or 'per'.")      
    
    #Check inputs
    if dollar_type not in ['GDP', 'MIL']:
        raise ValueError("Please set data_type to 'GDP', 'POP' or 'MIL'")
    
    #Establish a blank dataframe with identical columns for capturing the total values
    bottom=df_val.iloc[0,:]*0
    
    #Initialize plot
    plt.figure(figsize=(11,4))
    for i in range(10):
        #Plot the values for that country, with bottom = the sum total of all countries before it
        plt.bar(df_val.columns[:], df_val.iloc[i,:], bottom=bottom)
        #Add the values for that country to the sum total dataframe
        bottom=df_val.iloc[i,:]+bottom

    #Assign titles and Axis labels
    plt.title(str_dict[dollar_type][0])
    plt.ylabel(str_dict[dollar_type][1])
    plt.legend(df_dollars.iloc[:,0], bbox_to_anchor=(1.0, .9))
    
    plt.show()

```

```python

global_stack_dollars(data_gdp, data_pop, 'GDP', 'abs')
global_stack_dollars(data_gdp, data_pop, 'GDP', 'per')

global_stack_dollars(data_mil, data_pop, 'MIL', 'abs')
global_stack_dollars(data_mil, data_pop, 'MIL', 'per')
```

In the upper plots, we can see that global GDP was steadily rising up to 2015, driven largely by China's growing economy. There was a slight dip in 2015, then further growth.

```python

def changes_in_mil_spending(df_mil, year_start=2010, year_end=2017, country='all', mode='abs', span='annual', plot=True):
    """
    Plots the changes in military spending, either annually or comparing two points in time. 
    Can show all countries, or a reduced selection.
    
    Inputs:
    df_mil : The dataframe to be plotted, containing Military data
    year_start: The first year to be plotted, must be 2010 or later. If 2010, the first column plotted will be 2001.
    year_end: The last year to be plotted, must be 2017 or earlier. 
    country: The list of countries to be plotted. Default is 'all' to plot all countries.
    mode: Changes presented as change in absolute amount ('abs'), or as percentage change from the previous year ('per').
    span: Timeframe to compare, can show each year ('annual') or compare ony the two endpoints ('ends').
    plot: Deaults to True, will print plot. If False, will return dataframe of the changes in spending instead.
    """

    #Input checking
    if year_start<2010:
        raise ValueError("Please set year_start to 2010 or later") 
        
    if year_end>2017:
        raise ValueError("Please set year_end to 2017 or earlier") 
        
    if mode=='abs':
        str_add='in Billions (USD)' 
    elif mode=='per':
        str_add='in Percent Change'
    else:
        raise ValueError("Please set mode to 'abs' or 'per'.") 
        
    #Year Filter
    if span == 'annual':
        #Create a list of all years between start and end to keep
        years_keep=[str(i) for i in range(year_start-1, year_end+1)]

    elif span == 'ends':  
        #Create a list of the start and end years to keep
        years_keep=[str(year_start), str(year_end)]

    else:
        raise ValueError('Please set span to "annual" or "ends".')
    
    #Filter by the ID columns and selected years, dropping all other columns
    years_keep=['Country Name', 'Country Code']+years_keep
    df_mil=df_mil.filter(years_keep, axis=1)
    
    
    #Country filter
    if country !='all':
        #If multiple countries in a list, filter for all
        if type(country)==list:
            df_mil=df_mil[df_mil['Country Name'].isin(country)]
        #If only a single country, filter for it alone
        else:
            df_mil=df_mil[df_mil['Country Name'] == country ]

    #Initialize dataframe for tracking changes in spending with identical columns, and set values to 0
    deltas=df_mil.copy()
    deltas.iloc[:,2:]=0
    
    #Calculate change in spending
    #Selects the indices of the columns of the now shortened dataframe, starting after the ID columns
    for year in range(3, df_mil.shape[1]):
        if mode.lower()=='abs': #Absolute spending 
            deltas.iloc[:, year]=df_mil.iloc[:, year]-df_mil.iloc[:, year-1]
        elif mode.lower()=='per': #Percent change in spending 
            deltas.iloc[:, year]=(df_mil.iloc[:, year]-df_mil.iloc[:, year-1])/df_mil.iloc[:, year-1]*100
        else:
            raise ValueError('Please set mode to "abs" or "per".')
    
    #Transform dataframe from wide to tall using melt       
    deltas.drop(deltas.columns[2], axis=1, inplace=True) #Drop the first year (base before deltas start)
    melt_deltas=pd.melt(deltas, id_vars=['Country Name', 'Country Code'])
    
    #Checks if the plot is required, else returns dataframe
    if plot==False:
        return melt_deltas
    
    
    
    #Plot the data
    #Initialize Dataframe
    fig, ax= plt.subplots(figsize=(11,4))  
    #Use seaborn to make a grouped barplot showing each country grouped by year.
    #The y-value is he change in military spending calculated above
    sns.barplot(ax=ax, data=melt_deltas, hue='Country Name', x='variable', y='value',)
    
    #Plot a horizontal line at the x-axis for ease of viewing
    ax.axhline(y=0, color='k')
    
    #Add legend, titles and axis labels
    plt.legend(bbox_to_anchor=(1.0, .9))
    plt.ylabel(f'Change in Military Expenditure{str_add}')
    plt.xlabel('')
    plt.title(f'Changes in Military Spending from {year_start} to {year_end} {str_add}')
    
    #If comparing 2 end points, don't show the year as x-tick
    if span=='ends':
        ax.set_xticks([])
        

    plt.show()
```

```python
#Looking at all countries, in absolute change    
changes_in_mil_spending(data_mil, year_start=2010, year_end=2017, country='all')
changes_in_mil_spending(data_mil, year_start=2010, year_end=2017, country='all', span='ends')

#Looking at the countries identified above to be fluctuating  
#changes_in_mil_spending(data_mil, year_start=2010, year_end=2017, country=['Saudi Arabia', 'China', 'Russia'], span='annual')
```

```python
#Looking at all countries, in percent change    
changes_in_mil_spending(data_mil, year_start=2010, year_end=2017, country='all', mode='per')
changes_in_mil_spending(data_mil, year_start=2010, year_end=2017, country='all', mode='per',span='ends')

#Looking at the countries identified above to be fluctuating  
changes_in_mil_spending(data_mil, year_start=2010, year_end=2017, country=['Saudi Arabia', 'China', 'Russia'], mode='per',span='annual')
```

** Findings **

Most of the countries identified (Germany, France, UK, Iran, Israel, Italy ) have had relatively little change in their military spending between 2010 and 2017, in terms of both absolute and percent change. 

Some countries (China, Saudi Arabia and South Korea) have had an increase in overall spending in this same timespan.

While Russia had fluctuated, their 2017 military expenditure is roughly equal to their 2010 value, though this may be a coincidence rather than a trend.  

```python

def global_map_military_spending(df_mil, year_start=2010, year_end=2017, mode='per'):
    """
    Plots the changes in military spending between two points in time as a cartopy map. 
    Changes in spending are calculated using the changes_in_mil_spending() function.
    
    Inputs:
    df_mil : The dataframe to be plotted, containing Military data
    year_start: The first year to be plotted, must be 2010 or later. 
    year_end: The last year to be plotted, must be 2017 or earlier. 
    mode: Changes presented as change in absolute amount ('abs'), or as percentage change from the previous year ('per').
   
    """
    
    #First, calculate the changes in spending using the previously defined changes_in_mil_spending() function.
    gradient_df=changes_in_mil_spending(df_mil, year_start=year_start,year_end=year_end, mode=mode,
                                          span='ends', country='all', plot=False)
    
    
    #Calculate the RBGA Values for each country's returned values
    
    #Find the max change from 0, either positive or negative
    max_gradient=np.array([abs(gradient_df.value.min()), abs(gradient_df.value.max())]).max()
    
    #Create the normalized gradient colorbar centered on the zero between the max and negative max
    #Then output a color for each value
    norm = Normalize(vmin=-max_gradient, vmax=max_gradient)
    color_vals=[cm.coolwarm(norm(val),) for val in gradient_df.value ]
    
    #Filter the gradient data to remove unneeded columns, and replace the values with the RGBA from above
    gradient_df=gradient_df.filter(['Country Name','value'])
    gradient_df['value']=color_vals
    
    #Make a list of countries for ease of reference
    country_list=gradient_df.iloc[:,0].tolist()
    
    #Load the appropriate global shapefiles and reader
    shpfilename = shpreader.natural_earth(resolution='110m',
                                          category='cultural',
                                          name='admin_0_countries')
    
    reader = shpreader.Reader(shpfilename)
    countries = reader.records()
    country = next(countries)
    
    
    #Initialize the figure
    fig=plt.figure(figsize=(14, 10))
    
    #Load the map
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.set_extent([180, -22, 0, 80],  crs=ccrs.PlateCarree())
    ax.add_feature(cartopy.feature.OCEAN,edgecolor='black')
    
    
    #Loop through the countries in the reader, and add to map if relevant
    for country in countries:
        if country.attributes['NAME'] in country_list :
            #Set the facecolour to our calculated value
            colour_calc=gradient_df[gradient_df['Country Name']==country.attributes['NAME']].iloc[0,1]
            ax.add_geometries(country.geometry, ccrs.PlateCarree(), facecolor=colour_calc, edgecolor='black')
            #Add label text near the center of each country (offset for legibility)
            x = country.geometry.centroid.x        
            y = country.geometry.centroid.y
            ax.text(x, y+2, country.attributes['NAME'], color='White', size=10, ha='center', va='center', transform=ccrs.PlateCarree(), 
                    path_effects=[PathEffects.withStroke(linewidth=3, foreground="k", alpha=.8)])
       
   
    #Add an axis for Israel - It needs a zoomed-in window or you can't see it.
    ax2 = fig.add_axes([0.725, 0.3, 0.16, 0.16], projection=ccrs.PlateCarree(central_longitude=34), autoscale_on=False)
    ax2.set_extent([29, 40, 27, 35],  crs=ccrs.PlateCarree())
    ax2.add_feature(cartopy.feature.OCEAN,edgecolor='black')
    
    #Need to reload the reader, as it does not loop around
    countries = reader.records()
    country = next(countries)
    
    #Loop to israel, and repeat the above process, but in the zoomed window
    for country in countries:
        if country.attributes['NAME'] == "Israel":
            colour_calc=gradient_df[gradient_df['Country Name']=='Israel'].iloc[0,1]
            ax2.add_geometries(country.geometry, ccrs.PlateCarree(), facecolor=colour_calc, edgecolor='black')
            x = country.geometry.centroid.x        
            y = country.geometry.centroid.y
            ax2.text(x+2.4, y, country.attributes['NAME'], color='White', size=11, ha='center', va='center', transform=ccrs.PlateCarree(), 
                    path_effects=[PathEffects.withStroke(linewidth=3, foreground="k", alpha=.8)])
       
    
    #Add stand-alone colourbar to show the direction of the gradient
    cbar_ax = fig.add_axes([0.91, 0.33, 0.01, 0.36])
    cbar_ax.axes.get_xaxis().set_visible(False)
    cbar_ax.axes.get_yaxis().set_visible(False)
    ColorbarBase(cbar_ax, cmap='coolwarm', orientation = 'vertical')
    
    
    #Set the title
    if mode=='per':
        str_add='Percent'
    else:
        str_add='Absolute'
    
    ax.set_title(f'{str_add} Change in Military Expenditure from {year_start} to {year_end}', fontsize=16)
    
    plt.show()
```

```python

global_map_military_spending(data_mil, year_start=2010, year_end=2017, mode='per')
global_map_military_spending(data_mil, year_start=2010, year_end=2017, mode='abs')


global_map_military_spending(data_mil, year_start=2015, year_end=2017, mode='per')
global_map_military_spending(data_mil, year_start=2015, year_end=2017, mode='abs')



```
